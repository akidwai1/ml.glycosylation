```python
import requests
import json

search_url = "https://api.glygen.org/supersearch/search"

payload = {
  "concept_query_list": [
    {
      "concept": "site",
      "query": {
        "aggregator": "$and",
        "aggregated_list": [],
        "unaggregated_list": [
          {
            "path": "glycosylation_flag",
            "order": 0,
            "operator": "$eq",
            "string_value": "true"
          }
        ]
      }
    }
  ]
}
   # the payload is the query for supersearch search

search_response = requests.post(search_url,json=payload)
print(search_response.text)  # printing out the json result
```

```python
list_data = search_response.json() # convert response to json file
list_id = list_data["results_summary"]["site"]["list_id"] # obatining id

list_url = "https://api.glygen.org/supersearch/list/"
payload = {
 "id": list_id

}
list_response = requests.post(list_url,json=payload)
print(list_response.text)  # printing out the json result
```

```python
import requests

filtered_list_taxid = []

list_response= list_response.json()

for item in list_response["results"]:
    if item["tax_id"] == 9606:
        filtered_list_taxid.append(item)

print(filtered_list_taxid)
```

```python
# code check for S-linked and C-linked glycosylation sites in the protein descriptions for tax ID 9606.
filtered_list_remove = []

for item in list_response["results"]:
    if item["tax_id"] == 9606 and "S-linked glycosylation" not in item.get("S-linked", "") and "C-linked glycosylation" not in item.get("C-linked", ""):
        filtered_list_remove.append(item)
print(filtered_list_remove)
```

```python
import pandas as pd

df = pd.read_csv("human_protein_go_annotation.csv")

protein_accession = input("Enter the Protein Accession: ")

protein_df = df[df['uniprotkb_canonical_ac'] == protein_accession]

categories = ["molecular_function", "biological_process", "cellular_component"]

for category in categories:
    go_terms = protein_df.loc[protein_df['go_term_category'] == category, ['go_term_label', 'go_term_id']]
    print("\n")
    print(f"{category.replace('_', ' ').capitalize()}:")
    print("\n")
    formatted_go_terms = []
    for index, row in go_terms.iterrows():
        formatted_go_terms.append(f"{row['go_term_label']} ({row['go_term_id']})")
    print("; ".join(formatted_go_terms))
```

```python
categories = ["molecular_function", "biological_process", "cellular_component"] #categories to be included
#list to include values in each categories
for protein_accession in df1['uniprotkb_canonical_ac']: #for each protein in the original data set
    biological_list = []
    molecular_list = []
    cellular_list = []
    protein_df = df[df['uniprotkb_canonical_ac'] == protein_accession]
    for category in categories: #extracting info based on categories into the corresponsing list
        go_terms = protein_df.loc[protein_df['go_term_category'] == category, ['go_term_label', 'go_term_id']]
        if category == "molecular_function":
            for index, row in go_terms.iterrows():
                molecular_list.append(f"{row['go_term_label']} ({row['go_term_id']})")
        elif category == "biological_process":
            for index, row in go_terms.iterrows():
                biological_list.append(f"{row['go_term_label']} ({row['go_term_id']})")
        elif category == "cellular_component":
            for index, row in go_terms.iterrows():
                cellular_list.append(f"{row['go_term_label']} ({row['go_term_id']})")
    biological_process = ', '.join(map(str, biological_list)) #converting to string and adding ','
    molecular_function = ', '.join(map(str, molecular_list))
    cellular_component = ', '.join(map(str, cellular_list))
    df1.loc[df1.uniprotkb_canonical_ac == protein_accession,['molecular_function']] = molecular_function #adding info to the big dataset df2
    df1.loc[df1.uniprotkb_canonical_ac == protein_accession,['biological_process']] = biological_process
    df1.loc[df1.uniprotkb_canonical_ac == protein_accession,['cellular_component']] = cellular_component
```

```python
import requests  # step 9 “glytoucan_ac"

for i in df1['uniprotkb_canonical_ac']:
    response = requests.post(url=("https://api.glygen.org/protein/detail/" + i), json={"uniprot_canonical_ac": i})
    protein_data = response.json()

    gene_name = None
    for g in protein_data['gene_names']:
        if g['resource'] == 'UniProtKB' and g['type'] == 'recommended':
            gene_name = g['name']
            break
        elif g['resource'] == 'UniProtKB':
            gene_name = g['name']
            break

    glytoucan_ac = None
    for gly in protein_data["glycosylation"]:
        glytoucan_ac = gly.get('glytoucan_ac')

    df1.loc[df1.uniprotkb_canonical_ac == i, "glytoucan_ac"] = glytoucan_ac
    print(f"Glytoucan_ac for {i}: {glytoucan_ac}")

    df1.loc[df1.uniprotkb_canonical_ac == i, "gene_name"] = gene_name
    print(f"Gene name for {i}: {gene_name}")
```

```python
import requests

url = "https://api.glygen.org/glycan/detail/G49108TO/"
response = requests.get(url)
glycan_data = response.json()

subtypes = []
core_fucosylated = None
for classification in glycan_data.get("classification", []):
    subtype_name = classification.get("subtype", {}).get("name")
    if subtype_name:
        subtypes.append(subtype_name)
    if subtype_name == "Core-fucosylated":
        core_fucosylated = "Y"
    else:
        core_fucosylated ="N"

formatted_subtypes = ";".join(subtypes)
print(f"Extracted subtypes: {formatted_subtypes}")

df1["glycan_subtype"] = formatted_subtypes
df1["core_fucosylated"] = core_fucosylated

df1
```
